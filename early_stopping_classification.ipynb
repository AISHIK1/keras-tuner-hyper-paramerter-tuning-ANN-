```python
from sklearn.datasets import make_circles
```

```python
X,y=make_circles(n_samples=100,noise=0.1,random_state=1)
```

```python
import seaborn as sns
```

```python
X.shape
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y,style=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model=Sequential()
model.add(Dense(256,activation='relu',input_dim=2))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history=model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=900)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'],label='loss')
plt.plot(history.history['val_loss'],label='val_loss')
plt.legend()
```

```python
from mlxtend.plotting import plot_decision_regions
from matplotlib.colors import ListedColormap
```

```python
plot_decision_regions(X_test, y_test.ravel(), clf = model, legend=2)
plt.show()
```

```python
X,y=make_circles(n_samples=100,noise=0.1,random_state=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
model=Sequential()
model.add(Dense(256,activation='relu',input_dim=2))
model.add(Dropout(0.5))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping

callback=EarlyStopping(
    monitor="val_loss",
    min_delta=0.01,
    patience=50,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=900,callbacks=[callback])
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'],label='loss')
plt.plot(history.history['val_loss'],label='val_loss')
plt.legend()

plt.plot(history.history['accuracy'],label='loss')
plt.plot(history.history['val_accuracy'],label='val_loss')
plt.legend()
```

```python
plot_decision_regions(X_test, y_test.ravel(), clf = model, legend=2)
plt.show()
```
